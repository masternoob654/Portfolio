## Описание проекта

### Описание этапа разработки

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. *slab*, «плита»).

### Описание данных

Данные состоят из файлов, полученных из разных источников:

- `data_arc.csv` — данные об электродах;
- `data_bulk.csv` — данные о подаче сыпучих материалов (объём);
- `data_bulk_time.csv` *—* данные о подаче сыпучих материалов (время);
- `data_gas.csv` — данные о продувке сплава газом;
- `data_temp.csv` — результаты измерения температуры;
- `data_wire.csv` — данные о проволочных материалах (объём);
- `data_wire_time.csv` — данные о проволочных материалах (время).

Во всех файлах столбец `key` содержит номер партии. В файлах может быть несколько строк с одинаковым значением `key`: они соответствуют разным итерациям обработки.

## Предобработка данных

Для начала унифицируем названия столбцов.

```python
df_arc.columns = ['key', 'start_time','end_time', 'active_power', 'reactive_power']
df_bulk.columns = ['key', 'bulk_1', 'bulk_2', 'bulk_3', 'bulk_4', 'bulk_5', 'bulk_6', 'bulk_7', 
                   'bulk_8', 'bulk_9', 'bulk_10', 'bulk_11', 'bulk_12', 'bulk_13', 'bulk_14', 'bulk_15']
df_bulk_time.columns = ['key', 'bulk_1', 'bulk_2', 'bulk_3', 'bulk_4', 'bulk_5', 'bulk_6', 'bulk_7',
                        'bulk_8', 'bulk_9', 'bulk_10', 'bulk_11', 'bulk_12', 'bulk_13', 'bulk_14', 'bulk_15']
df_gas.columns = ['key', 'gas']
df_temp.columns = ['key', 'time', 'temperature']
df_wire.columns = ['key', 'wire_1', 'wire_2', 'wire_3', 'wire_4', 'wire_5', 'wire_6', 'wire_7', 'wire_8', 'wire_9']
df_wire_time.columns = ['key', 'wire_1', 'wire_2', 'wire_3', 'wire_4', 'wire_5', 'wire_6', 'wire_7', 'wire_8', 'wire_9']
```

```python
df_wire = df_wire.fillna(0)
df_bulk = df_bulk.fillna(0)
```

От аномальных значений температуры избавимся при удалении промежуточных значений, а пока высчитаем дополнительные переменные на основе текущих - время нагрева, общую мощность и итоговую работу.

```python
df_arc['time'] = df_arc['end_time'] - df_arc['start_time'] # время нагрева
df_arc['time'] = df_arc['time'].dt.seconds # перевод минут в секунды

df_arc['sum_power'] = np.sqrt(df_arc['active_power'] ** 2 + df_arc['reactive_power'] ** 2) # полная мощность

df_arc['total_work'] = df_arc['sum_power'] * df_arc['time'] # полная работа
```

Следующим шагом будет объединение датасетов в один для последующего обучения на нём модели, но прежде всего нужно оставить ключи, которые есть для первого и последнего замеров температуры.

```python
df_temp = df_temp.dropna(subset = ['temperature']).reset_index(drop = True)
df_temp = df_temp.sort_values(by = ['time'])

count = df_temp.groupby(by = 'key', as_index = False).count()
condition = count[count['time'] == 1]['key'].tolist()
df_temp = df_temp[~df_temp['key'].isin(condition)]
```

```python
df_temp = df_temp.groupby(by ='key').agg(['first', 'last']).reset_index()
df_temp.columns = df_temp.columns.droplevel()
df_temp.columns = ['key', 'measurement_time_1', 'measurement_time_2', 'first_temp', 'last_temp']
df_temp = df_temp[['key', 'first_temp', 'last_temp']]
```

```python
df_temp
```

Теперь можно приступить к объединению факторов из разных датасетов в один. Объединять будем на основе общих значениях ключей.

### Объединение датафреймов

Группируем по таблицу по ключам, оставляем сумму значений полной работы (остальные значения мы отбрасываем, так как только полную работу можно суммировать)

```python
df_arc_iter = df_arc.groupby(by = 'key', as_index = False).agg(total_work = ('total_work', 'sum'))
```

```python
df_arc_iter.head()
```

```python
final_df = (df_temp.merge(df_arc_iter, how = 'inner', on = 'key')
                 .merge(df_bulk, how = 'inner', on = 'key')
                 .merge(df_wire, how = 'inner', on = 'key')
                 .merge(df_gas, how = 'inner', on = 'key'))
```

```python
final_df.info()
```

```python
final_df.head()
```

Критичных выбросов нет как таковых, оставим целевой признак без изменений. Теперь построим матрицу корреляции для признаков в общем датафрейме.

Видим отсутствие корреляции между целевым признаком и признаками `wire_5`, `wire_7`. Это как раз те признаки, которые содержали малое количество значений. Удалим все эти признаки кроме `total_work`, при этом можно удалить столбец `key`, так как он содержит уникальные значения и нужен был для отсечения тех строк, которых нет в таблице со значениями целевого признака `last_temp`.

```python
final_df = final_df.drop(['key', 'wire_5', 'wire_7'], axis = 1)
```

## Обучение модели

Будем проверять три модели: LGBMRegressor, случайный лес и линейную регрессию. Сразу инициализируем простейшую модель для проверки нашей будущей модели.

```python
%%time 

dummy_model = DummyRegressor(strategy = "mean")
score_dummy = cross_val_score(dummy_model, features_train, target_train, scoring = 'neg_mean_absolute_error')
print('MAE простейшей модели равен', max(score_dummy) * -1)
```

Таким образом, 7.2 - значение, с которым мы будем сравнивать результаты наших моделей.

### LGBMRegressor

```python
%%time

lgbm = LGBMRegressor() 
lgbm_params = [{'num_leaves':[x for x in range(10, 25, 2)], 
                'learning_rate':[0.1],
                'random_state':[RANDOM_STATE]}]

gscv_lgbm = GridSearchCV(lgbm, lgbm_params, scoring = 'neg_mean_absolute_error', n_jobs = -1, verbose = 10)
gscv_lgbm.fit(features_train, target_train)

print('Лучшие гиперпараметры:', gscv_lgbm.best_params_, '\n')

mae_lgbm = (max(gscv_lgbm.cv_results_['mean_test_score']) * -1)
print('MAE для LGBMRegressor равен', mae_lgbm)
print()
```

### Случайный лес

```python
%%time

rfr = RandomForestRegressor() 
rfr_params = [{'criterion':['mae'],
               'n_estimators':[x for x in range(100, 501, 50)],
               'random_state':[RANDOM_STATE]}]

gscv_rfr = GridSearchCV(rfr, rfr_params, scoring = 'neg_mean_absolute_error', n_jobs = -1, verbose = 10)
gscv_rfr.fit(features_train, target_train)

print('Лучшие гиперпараметры:', gscv_rfr.best_params_, '\n')

mae_rfr = max(gscv_rfr.cv_results_['mean_test_score']) * -1
print('MAE для случайного леса равен', mae_rfr)
```

### Линейная регрессия

```python
%%time

lr = LinearRegression()
lr_params = {}

grid_lr = GridSearchCV(lr, param_grid=lr_params, scoring = 'neg_mean_absolute_error')
grid_lr.fit(features_train, target_train)
predict_train = grid_lr.predict(features_train)

mae_lr = mean_absolute_error(target_train, predict_train)

print('MAE для линейной регрессии равен', mae_lr)
print()
```

```python
indices = ['LGBMRegressor',
           'Случайный лес',
           'Линейная регрессия']

data = {'MAE на обучающей выборке': [mae_lgbm,
                                     mae_rfr,
                                     mae_lr]}

df_models = pd.DataFrame(data = data, index = indices)

df_models.sort_values(by = 'MAE на обучающей выборке')
```
